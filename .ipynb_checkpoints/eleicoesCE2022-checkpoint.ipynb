```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('candidatos_ce.csv', header=None, encoding = "ISO-8859-1", delimiter=';')
display(df)
```

```python
df.columns = df.iloc[0]
df = df.drop(0)
df.head(10)
```

```python
# Excluir colunas desnecessarias
df = df.drop(columns=['DT_GERACAO','HH_GERACAO','CD_TIPO_ELEICAO','NM_TIPO_ELEICAO','NR_TURNO','CD_ELEICAO','DS_ELEICAO','DT_ELEICAO','TP_ABRANGENCIA','SG_UF','SG_UE','NM_UE','SQ_CANDIDATO','NM_CANDIDATO','NM_SOCIAL_CANDIDATO','NR_PROCESSO','NM_TIPO_DESTINACAO_VOTOS','NR_CPF_CANDIDATO','NM_EMAIL','CD_SITUACAO_CANDIDATURA','ST_DECLARAR_BENS','NR_PROTOCOLO_CANDIDATURA','CD_SITUACAO_CANDIDATO_PLEITO','CD_SITUACAO_CANDIDATO_URNA','ST_CANDIDATO_INSERIDO_URNA','CD_SITUACAO_CANDIDATO_TOT','DS_SITUACAO_CANDIDATO_TOT','ST_PREST_CONTAS','TP_AGREMIACAO','VR_DESPESA_MAX_CAMPANHA','CD_SIT_TOT_TURNO','DS_SIT_TOT_TURNO','DS_SITUACAO_CANDIDATO_PLEITO','DS_SITUACAO_CANDIDATO_URNA','NM_PARTIDO','CD_COR_RACA','NR_FEDERACAO','CD_DETALHE_SITUACAO_CAND','DS_DETALHE_SITUACAO_CAND','NM_FEDERACAO','SG_FEDERACAO','DS_COMPOSICAO_FEDERACAO','SQ_COLIGACAO','NM_COLIGACAO','DS_COMPOSICAO_COLIGACAO','CD_NACIONALIDADE','DS_NACIONALIDADE','ST_REELEICAO','NR_TITULO_ELEITORAL_CANDIDATO'])
df.head()
```

```python
# Alterar nome das colunas
df.rename(columns={
    'ANO_ELEICAO':'Ano Eleição', 
    'CD_CARGO':'ID Cargo',
    'DS_CARGO':'Cargo',
    'NR_CANDIDATO':'Número Candidato',
    'NM_URNA_CANDIDATO':'Nome Candidato',
    'DS_SITUACAO_CANDIDATURA':'Situação',
    'NR_PARTIDO':'Número Partido',
    'SG_PARTIDO':'Sigla Partido',
    'SG_UF_NASCIMENTO':'UF Nascimento',
    'CD_MUNICIPIO_NASCIMENTO':'ID Mun Nasc',
    'NM_MUNICIPIO_NASCIMENTO':'Municipio',
    'NR_IDADE_DATA_POSSE':'Idade Candidato',
    'CD_GENERO':'ID Genero',
    'DS_GENERO':'Genero',
    'CD_GRAU_INSTRUCAO':'ID Grau Instrução',
    'DS_GRAU_INSTRUCAO':'Grau Instrução',
    'CD_ESTADO_CIVIL':'ID Estado Cívil',
    'DS_ESTADO_CIVIL':'Estado Civil',
    'DS_COR_RACA':'Raça',
    'CD_OCUPACAO':'ID Ocupação',
    'DS_OCUPACAO':'Ocupação'
}, inplace=True)

df.head()
```

```python
# Conversão de valores objeto para inteiro
columns = ['Ano Eleição','ID Cargo','Número Candidato','Número Partido','ID Mun Nasc','Idade Candidato','ID Genero','ID Grau Instrução','ID Estado Cívil','ID Ocupação']
df[columns] = df[columns].apply(lambda x: x.str.replace(',', '.').astype('int'))
```

```python
#analisar campos nulos
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull());
```

```python
# Análise da idade dos candidatos na eleição de 2022
plt.figure(figsize=(12,6))
plt.title('Analisando a idade dos candidatos', loc='left', fontsize=14)
plt.style.use('ggplot')

sns.barplot(data=df, x='Idade Candidato', y='Ano Eleição', ci=None, color='#11e2e3', estimator=sum);
```
